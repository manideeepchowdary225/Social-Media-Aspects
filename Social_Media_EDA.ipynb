```python
SM.groupby('How many hours per day do you spend on social media? ').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
SM.groupby('What is your age group? ').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
SM.groupby('  What is your occupation? ').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['How has social media affected your real-life relationships? '].value_counts()
    for x_label, grp in SM.groupby('How many hours per day do you spend on social media? ')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('How many hours per day do you spend on social media? ')
_ = plt.ylabel('How has social media affected your real-life relationships? ')
```

```python
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Where do you live? '].value_counts()
    for x_label, grp in SM.groupby('  What is your occupation? ')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('  What is your occupation? ')
_ = plt.ylabel('Where do you live? ')
```

```python
# Create a boxplot for 'How many hours per day do you spend on social media?' by age group
plt.figure(figsize=(12, 6))
sns.boxplot(x='What is your age group? ', y='How many hours per day do you spend on social media? ', data=SM, palette='Set3')
plt.title('Social Media Usage Time by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Hours per Day')
plt.show()

# Create a violinplot for 'How has social media affected your real-life relationships?' by occupation
plt.figure(figsize=(14, 8))
sns.violinplot(x='  What is your occupation? ', y='How has social media affected your real-life relationships? ', data=SM, palette='pastel')
plt.xticks(rotation=45, ha='right')
plt.title('Impact of Social Media on Relationships by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Impact on Relationships')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='How many hours per day do you spend on social media? ', data=SM)
plt.title('Distribution of Social Media Usage Time')
plt.xlabel('Hours per Day')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=SM['Do you feel that social media affects your mental health?  '], palette='Reds')
plt.xticks(rotation=50)
plt.show()
```

```python
SM['Which social media platforms do you use the most?  '].value_counts().plot(kind='pie', figsize=(10,6), autopct='%.2f', shadow=True)
plt.show()
```

```python
SM['How often do you use WhatsApp for communication? '].value_counts().plot(kind='pie',
                                                                            figsize=(10,6), autopct='%.2f', shadow=True,
                                                                           colors=['purple', 'steelblue', 'tomato'])
plt.show()
```

```python
SM['What is your age group? '].value_counts().plot(kind='pie',
                                                                            figsize=(10,6), autopct='%.2f', shadow=True,
                                                                           colors=['steelblue', 'tomato'])
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=SM['Where do you live? '], palette='magma',hue=SM['How often do you use WhatsApp for communication? '])
plt.xticks(rotation=50)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=SM['Where do you live? '],hue=SM['Which social media platforms do you use the most?  '], data=SM, palette='winter')
plt.xticks(rotation=50)
plt.legend()
plt.show()
```
